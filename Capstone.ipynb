```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import csv

genre = pd.read_csv("genre.csv")
user = pd.read_csv("users.csv")

data = pd.merge(genre, user, on="user_id")

data = data.query('age >= 10 & age < 25')

data.drop(['user_id'], axis=1, inplace=True)
data.drop(['country_y'], axis=1, inplace=True)
data.drop(['age'], axis=1, inplace=True)
data.drop(['gender'], axis=1, inplace=True)
data.drop(['registered_unixtime'], axis=1, inplace=True)
data.drop(['playcount'], axis=1, inplace=True)

data = data.sum(axis=0)
data = data.sort_values()

data.plot.bar(x="Genre", y="Playcount", title="Ages 10-25")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import csv
import pycountry_convert as pc

genre = pd.read_csv("genre.csv")
user = pd.read_csv("users.csv")

data = pd.merge(genre, user, on="user_id")

data = data.loc[data['country_y'] == 'US']

data.drop(['user_id'], axis=1, inplace=True)
data.drop(['country_y'], axis=1, inplace=True)
data.drop(['age'], axis=1, inplace=True)
data.drop(['gender'], axis=1, inplace=True)
data.drop(['registered_unixtime'], axis=1, inplace=True)
data.drop(['playcount'], axis=1, inplace=True)

data = data.idxmax(axis=1)
data = data.value_counts(ascending=True)

data.plot.bar(x="Prefered Genre", y="Number of people", title="United States" )
```
