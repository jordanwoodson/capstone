```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import csv

genre = pd.read_csv("genre.csv")
user = pd.read_csv("users.csv")

data = pd.merge(genre, user, on="user_id")

data = data.drop(data.columns[[0,21,22,23]],axis=1)
data = data.idxmax(axis=1)
data = data.value_counts(ascending=True)


data.plot.bar(x="Prefered Genre", y="Number of people", title="All ages" )
```
